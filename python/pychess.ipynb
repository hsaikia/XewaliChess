```python
G = nx.DiGraph()
node_colormap = []
for x in chess.SQUARES:
    Piece = board.piece_at(x)
    if Piece == None :
        node_colormap.append("red")
    elif Piece.color == chess.WHITE :
        node_colormap.append("#eeeeee")
    elif Piece.color == chess.BLACK :
        node_colormap.append("#333333")
            
    G.add_node(chess.SQUARE_NAMES[x])

for f in chess.SQUARES :
    atks = board.attacks(f)
    for sq in atks :
        G.add_edge(chess.SQUARE_NAMES[f], chess.SQUARE_NAMES[sq])
    
fig = plt.figure(figsize=(15, 15))
nx.draw(G, with_labels=True, layout='planar_layout', node_color = node_colormap, font_weight='bold', font_size=24, node_size=4000)
    
```
